```python
import numpy as np
from sklearn.linear_model import LinearRegression

# 1. 데이터 설정
X = np.array([[1], [2], [3], [4]])
y = np.array([3, 5, 7, 9])

# 2. 모델 학습
model = LinearRegression().fit(X, y)

# 3. 예측 (x=5일 때 정답 11.0)
result = model.predict([[5]])
print(f"예측 결과: {result[0]}")
```
